```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("dataa.csv", index_col=False)

# Drop the column named "Unnamed: 0"
data = data.drop(columns=["Unnamed: 0","ID"])
data['date'] = data['date'].str.split().str[0]
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le_count = 0

for col in data:
    if data[col].dtype == 'object':
        if len(list(data[col].unique())) <= 3:
            le.fit(data[col])
            data[col] = le.transform(data[col])
            le_count += 1


data['date'] = pd.to_datetime(data['date'])

data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day

data = data.drop(columns=['date'])

data = pd.get_dummies(data, columns=['campaign_type'])
data = pd.get_dummies(data, columns=['region'])
data = data.replace({True: 1, False: 0})
mapping = {'18-25': 1, '26-35': 2, '36-45': 3, '46-55': 4, '56+': 5}
data['customer_age_range'] = data['customer_age_range'].map(mapping)
```

```python
scaler = MinMaxScaler()
data['sales'] = scaler.fit_transform(data[['sales']])
```
